```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
"""
the difference between xy and xy1:
pd.read_csv reads from second row, because it recognizes first row as column_name
on the other hand, np.loadtxt reads from first row
"""
xy = np.loadtxt('./data/data-01-test-score.csv', delimiter=',', dtype=np.float32)
xy1 = np.array(pd.read_csv("./data/data-01-test-score.csv"), dtype='float32')
```

```python
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
```

```python
# Make sure the shape and data are OK
print(x_data, "\nx_data shape:", x_data.shape)
print('-'*30)
print(y_data, "\ny_data shape:", y_data.shape)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_dim=3))
model.add(tf.keras.layers.Activation('linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(lr=1e-5))
history = model.fit(x_data, y_data, epochs=2000)
```

```python
# Ask my score
print("Your score will be ", model.predict([[100, 70, 101]]))
print("Other scores will be ", model.predict([[60, 70, 110], [90, 100, 80]]))
```
