# xor can be explained by multi layer perceptron
we can use SGD, but for the similarity with 09-3. we use Adam optimizer

(Modified)
When you pass the
strings 'accuracy' or 'acc', we convert this to one of  
`tf.keras.metrics.binary_accuracy`,  
`tf.keras.metrics.categorical_accuracy`,  
`tf.keras.metrics.sparse_categorical_accuracy`  
based on the loss
function used and the model output shape.

```python
import numpy as np
import tensorflow as tf
```

```python
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype='float32')
y_data = np.array([[0], [1], [1], [0]], dtype='float32')
print(x_data)
print(y_data)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=2, input_dim=2))
model.add(tf.keras.layers.Activation('sigmoid'))
model.add(tf.keras.layers.Dense(units=1, input_dim=2))
model.add(tf.keras.layers.Activation('sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.Adam(lr=0.1), metrics=[tf.keras.metrics.binary_accuracy])
model.summary()
```

```python
history = model.fit(x_data, y_data, epochs=10000)
```

```python
predictions = model.predict(x_data)
print(f'predictions:\n{predictions}')
```

```python
score = model.evaluate(x_data, y_data)
print(f'accuracy: {score[1]}')
```
