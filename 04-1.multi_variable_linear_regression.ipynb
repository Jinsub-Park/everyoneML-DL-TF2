```python
# basis(libraries and data)
import tensorflow as tf
import numpy as np

x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_dim=3))  # unit is one and x is a 3-fold vector
model.add(tf.keras.layers.Activation('linear'))  # can be omitted since it is a default setting.
```

```python
model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(lr=1e-5))
model.summary()
```

```python
history = model.fit(x_data, y_data, epochs=1000)
```

```python
y_predict = model.predict(np.array([[72., 93., 90.]]))
print(y_predict)
```
