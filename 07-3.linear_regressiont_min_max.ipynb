```python
import tensorflow as tf
import numpy as np
```

```python
xy = np.array([[828.659973, 833.450012, 908100, 828.349976, 831.659973],
               [823.02002, 828.070007, 1828100, 821.655029, 828.070007],
               [819.929993, 824.400024, 1438100, 818.97998, 824.159973],
               [816, 820.958984, 1008100, 815.48999, 819.23999],
               [819.359985, 823, 1188100, 818.469971, 818.97998],
               [819, 823, 1198100, 816, 820.450012],
               [811.700012, 815.25, 1098100, 809.780029, 813.669983],
               [809.51001, 816.659973, 1398100, 804.539978, 809.559998]])
```

```python
def minMaxScaler(array):
    numerator = array - np.min(array, axis=0)
    denominator = np.max(array, axis=0) - np.min(array, axis=0)
    return numerator / (denominator + 1e-8)
```

```python
xy = minMaxScaler(xy)
x = xy[:, :-1]
y = xy[:, [-1]]
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_dim=4 , activation='linear')])
model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(lr=1e-1))
model.summary()
```

```python
history = model.fit(x, y, epochs=700)
```

```python
predictions = model.predict(x)
score = model.evaluate(x, y)
```

```python
print('Prediction:')
for i in range(len(y)):
    print('y_predict: {}, y_real: {}'.format(predictions[i], y[i]))
print('Cost: ', score)
```
